# 7. Fix Data Types

```python
# Convert categorical columns to 'Category' data type for better memory efficiency and faster analysis.
cat_cols = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

## 8. Dropping Irrelevant Columns

Some columns do not provide useful information for employee attrition analysis:

- **Over18 - contains** : only one category ("Y") for all employees.
- **StandardHours** : has the same value for every record.

These columns are removed to simplify the dataset and focus on meaningful features.

```python
# List of columns that are not useful for attrition analysis
drop_cols = ['Over18', 'StandardHours']

# Remove the selected columns from the dataset
df.drop(columns = drop_cols, inplace=True)
```

## 9. Creating Derived Features

To make the analysis more meaningful, new features are created from existing columns:

- **Age_Group**: Groups employees based on their age.
- **Income_Band**: Groups employees based on their monthly income.
- **Tenure_Band**: Groups employees based on how long they have worked in the company.
- **Is_Overtime**: Converts overtime values from Yes/No to 1/0.

These new columns help compare attrition across different age groups, income levels, work experience, and overtime status.

```python
# Create age groups to analyze attrition across different age ranges

df['Age_Group'] = pd.cut(df['Age'],
                         bins = [0, 25, 35, 45, 60],
                         labels = ['18-25', '26-35', '36-45', '45-60'])

# Divide employees into four income bands based on income distribution

df['Income_Band'] = pd.qcut(df['MonthlyIncome'],
                            q = 4,
                            labels = ['Low', 'Medium', 'High', 'Very High'])

# Group employees by years spent at the company

df['Tenure_Band'] = pd.cut(df['YearsAtCompany'],
                          bins = [-1, 2, 5, 10, 40],
                          labels = ['0-2', '3-5', '6-10', '10+'])

# Convert overtime status from Yes/No to numeric values (1/0)

df['Is_Overtime'] = df['OverTime'].map({'Yes':1, 'No':0})
```

## 10. Saving the Cleaned Dataset

After cleaning and preparing the data, the dataset is saved as a new CSV file.

This allows us to keep a separate copy of the processed data and use it later for analysis, visualization, or machine learning without repeating the data cleaning steps.

```python
# Save the cleaned dataset as a new CSV file without row index
df.to_csv('cleaned_hr_attrition.csv', index=False)

print("Cleaned dataset saved successfully as 'cleaned_hr_attrition.csv'")
```
